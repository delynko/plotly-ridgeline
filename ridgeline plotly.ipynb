```python
import pandas as pd
import plotly.graph_objects as go
from plotly.colors import n_colors

df = pd.read_csv(r"Z:\Dev\ridgeline\data\Lot_Spot_Occupancy_Chart02-10-2021_9-45-39am.csv")
df['datetime'] = pd.to_datetime(df['time'], unit='ms').dt.tz_localize('UTC').dt.tz_convert('America/Denver')
df['month'] = df['datetime'].dt.month
df['month_name'] = df['datetime'].dt.month_name()
df['hour'] = df['datetime'].dt.hour
df['day'] = df['datetime'].dt.day_name()
df['dow'] = df['datetime'].dt.dayofweek
fdf = df.where(df['hour']>5)
pdf = fdf.where(fdf['hour']<21).dropna()
adf = pdf.groupby(['month', 'month_name', 'day', 'dow']).agg({'count':'mean'}).sort_values(by=['month', 'dow']).reset_index()
```

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
months = reversed(range(1,13))
# months = reversed(months1)
colors = n_colors('rgb(5, 0, 200)', 'rgb(200, 10, 10)', 12, colortype='rgb')


fig = go.Figure()
for m, c in zip(months, colors):
    fig.add_trace(
        go.Violin(
            y=adf['month'][ adf['month'] == m ],
            x=adf['count'][ adf['month'] == m ],
            legendgroup=m,
            scalegroup=m,
            name=m,
            line_color='white',
            fillcolor=c
        )
    )

fig.update_traces(box_visible=False, meanline_visible=True, orientation='h', side='positive', width=4, points=False)
fig.update_layout(xaxis_showgrid=False, xaxis_zeroline=False, showlegend=False)
fig.update_layout(violinmode='group')
fig.show()
fig.write_html(r"Z:\Dev\ridgeline\html\month.html")
```
